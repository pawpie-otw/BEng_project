```python
import pandas as pd
import numpy as np
from random import choices
```

```python
files = ["2020_gry_zespolowe.xlsx", "2020_pozostale.xlsx","2020_spoty_walki.xlsx","2020_spoty_wodne.xlsx","2020_spoty_zimowe.xlsx"]
```

```python
df_s = [pd.read_excel(file, "TABLICA")
    for file in files]
```

```python
df_s[0]
```

```python
juniors_in_voivodeship=pd.DataFrame(np.array([df_s[i].loc[4:,df_s[i].columns[3::2]].sum(axis=1) 
for i in range(len(df_s))]).T).sum(axis=1)
juniors_in_voivodeship
```

```python
generally_in_voivodeship=pd.DataFrame(np.array([df_s[i].loc[4:,df_s[i].columns[2::2]].sum(axis=1) 
for i in range(len(df_s))]).T).sum(axis=1)
generally_in_voivodeship
```

```python
x=df_s[0]["Nazwa"][4:]
x.reset_index(inplace=True, drop=True)

result_df = pd.DataFrame({"voivodeship":x,
             "juniors": juniors_in_voivodeship,
              "seniors":generally_in_voivodeship-juniors_in_voivodeship})
result_df
```

```python
df_males = pd.read_csv("../people/voivodship_males_by_sexage.csv", index_col='index')
df_females = pd.read_csv("../people/voivodship_females_by_sexage.csv", index_col='index')
```

```python
m_add_fem_15_17_df = df_females[15:18].sum(axis=0)+df_males[15:18].sum(axis=0)
m_add_fem_15_17_df.name = "population 15-17yo"
```

```python
m_add_fem_18_40_df = df_females[17:41].sum(axis=0)+df_males[17:41].sum(axis=0)
m_add_fem_18_40_df.name = "population 18-40yo"
```

```python
merged_df = result_df.merge(pd.DataFrame(m_add_fem_15_17_df), left_on="voivodeship", right_index=True)
merged_df = merged_df.merge(pd.DataFrame(m_add_fem_18_40_df), left_on="voivodeship", right_index=True)
merged_df
```

```python
merged_df["junior_chance"]=merged_df.juniors/merged_df["population 15-17yo"]
merged_df["senior_chance"]=merged_df.seniors/merged_df["population 18-40yo"]
merged_df
```

```python
merged_df.set_index('voivodeship', drop=True, inplace=True)
merged_df
```
